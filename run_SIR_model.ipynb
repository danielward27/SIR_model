### Imports
All imports occur here:

```python
import numpy as np
from mesa import Agent, Model
from mesa.time import RandomActivation
from mesa.space import MultiGrid
from SIRModel import *
import matplotlib.pyplot as plt
import seaborn as sns
```

### Run model

```python
model = SIRModel(num_agents=500, initial_infected=0.02,
                 removed_probability=0.005, width=50, height=50)
for i in range(1000):
    model.step()
    
df = model.datacollector.get_model_vars_dataframe()
sns.lineplot(data=df)
```

### Social distancing
A simple way to approximate social distancing is to increase the size of the grid, whilst keeping everything else the same. Will this cause the flattening the curve behaviour?

```python
model = SIRModel(num_agents=500, initial_infected=0.02,
                 removed_probability=0.005, width=100, height=100)
for i in range(1000):
    model.step()
    
df = model.datacollector.get_model_vars_dataframe()
sns.lineplot(data=df)
```

Obviously this means very little for the real world as the model is so idealised, but it's still pretty cool!
